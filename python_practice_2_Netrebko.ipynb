9. Створіть дві множини. Продемонстуйте над ними операції: обʼєднання, різниці, пересічі та симетричної різниці. Використовуйте методи, що не змінюють множини, а створюють нові.

```python
a = {1, 2, 3,4}
b= {4, 5, 6,3}

union = a.union(b)
difference = a.difference(b)
intersection = a.intersection(b)
symmetric = a.symmetric_difference(b)

print("Об'єднання:", union)
print("Різниця:", difference)
print("Перетин:", intersection)
print("Симетрична різниця:", symmetric)
```

14. Створіть цілочисленну змінну зі своїм щасливим або улюбленим числом. Напишіть код, який би автоматизовував наступну задачу.

Якщо число парне, помножте його на 2 та виведіть результат на екран.

Але якщо число непарне, помножте його на 3 та додайте 1 і виведіть на екран.


```python
lucky_number = 12

if lucky_number % 2 == 0:
    lucky_number = lucky_number * 2
else:
    lucky_number = lucky_number * 3 + 1

print(lucky_number)
```

Якщо в імені є літери "o" та "n", Виведіть на екран повідомлення "O and N is in the name".

```python
name = "Karyna"
name = name.lower()
mylist = list(name)

count_O = mylist.count("o")
count_A = mylist.count("a")
count_N = mylist.count("n")
count_I = mylist.count("i")

print(count_N)
print(mylist)

if count_O != 0 and count_N != 0:
    print("O and N are in the name")

if count_A != 0 and count_I != 0:
    print("A and I are in the name")

if count_O != 0 and count_N == 0:
    print("Only O is in the name")

if count_O == 0 and count_N != 0:
    print("Only N is in the name")

if count_A != 0 and count_I == 0:
    print("Only A is in the name")

if count_A == 0 and count_I != 0:
    print("Only I is in the name")
```
